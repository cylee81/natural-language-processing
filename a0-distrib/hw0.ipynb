```python
from tokenizer import tokenize
from collections import Counter
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open('nyt.txt') as f:
    contents = f.read()
```

```python
res_simple = contents.split(" ")
rest_simple_dict = Counter(res_simple)
dict(sorted(rest_simple_dict.items(), key=lambda item: item[1], reverse=True))
```

```python
res = tokenize(contents)
```

```python
res.count()
```

```python
rest_dict = Counter(res)
```

```python
dict(sorted(rest_dict.items(), key=lambda item: item[1], reverse=True))
```

```python
np.linspace(1, len(rest_dict.values()), len(rest_dict.values()))
```

```python
plt.scatter(1/np.linspace(1, len(rest_dict.values()), len(rest_dict.values())), 
            sorted(list(rest_dict.values()), reverse=True))
plt.xlabel("reverse rank")
plt.ylabel("frequency")
plt.show()
```
